# Find paragraphs for a given content

```python
search_text = "Exposure"
search_text
```

```python
def find_paragraphs(doc, search_text, display_text=False, display_run_wise_text=False):
    found_in_paragraphs = []
    para_num = 0

    for p in doc.paragraphs:
        if p.text.find(search_text)>=0:
            found_in_paragraphs.extend([(para_num, p)])
            if display_text:
                print("Found in para_num =", para_num, "and text =", p.text)
                print("Total no of runs in this para =", len(p.runs))
                if display_run_wise_text:
                    for r in p.runs:
                        print(r.text)
                print("*-*-*")
        else:
            para_num += 1

    return found_in_paragraphs
```

```python
doc = Document(input_path)
print(len(doc.paragraphs))
```

```python
found_in_paragraphs = find_paragraphs(doc, search_text, True, True)
assert(len(found_in_paragraphs))
```

# Create section map

```python
def create_section_map(doc, display_text=False):
    section_map = {}
    rev_section_map = {}
    para_num = 0

    for key in section_texts:
        for p in doc.paragraphs[para_num:]:
            if p.text.find(section_texts[key])>=0:
                section_map[para_num] = section_texts[key]
                break
            else:
                para_num += 1

    assert(len(section_map)==len(section_texts))
    if display_text:
        print("section_map:", section_map)
    
    for key in section_map:
        rev_section_map[section_map[key]] = key

    if display_text:
        print("rev_section_map:", rev_section_map)
    
    return section_map, rev_section_map
```

```python
doc = Document(input_path)
print(len(doc.paragraphs))
```

```python
section_map, rev_section_map = create_section_map(doc, True)
```

# Find sections for a given content

```python
search_text = content_texts["op_find_sections"]
search_text
```

```python
doc = Document(input_path)
print(len(doc.paragraphs))
```

```python
def find_sections(doc, search_text):
    para_num = 0
    found_in_sections = []
    
    section_map, rev_section_map = create_section_map(doc)
    
    for p in doc.paragraphs[para_num:]:
        if p.text.find(search_text)>=0:
            for key in sorted(section_map.keys()):
                if para_num >= key:
                    found_in_sections.extend([(section_map[key], para_num)])
                    print("Found in section =", section_map[key], "in paragraph number =", para_num,
                          "and text =", p.text)
                    break
        else:
            para_num += 1
            
    return found_in_sections
```

```python
found_in_sections = find_sections(doc, search_text)
assert(len(found_in_sections))
```
